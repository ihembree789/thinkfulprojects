```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('QBStats_2016.csv')
```

```python
ax = sns.boxplot(y='cmp',data=df)  
plt.title('QB Completions per game')
sns.despine(offset=10, trim=False)
ax.set(xlabel='', ylabel='Completions')
plt.show()
```

Pro: The mean for completions is shown. Cons : It does not show the amount for completions. The nuance of distribution is missing. 

```python
cmp=df['cmp']
plt.hist(cmp, alpha=.5,data=df)  # alpha just controls the opacity
plt.xlabel('Completions')
plt.title('QB Completions per game')
plt.show()
```

Pros : This shows the amounts for completions and the nuance of distribution. Con: The mean for completions is missing.

```python
sns.distplot(cmp, hist=False, color="g", kde_kws={"shade": True})
plt.title('QB Completions per game')
plt.xlabel('Completions')
plt.show()
```

Pro : The distribution is shown with great detail. Cons: The mean for completions is missing. This does not show the amount for completions.

```python
sns.distplot(cmp, hist=True, color="b", kde_kws={"shade": True})
plt.title('QB Completions per game')
plt.xlabel('Completions')
plt.show()
```

Pro : The distribution is shown with great detail. Cons: The mean for completions is missing. This does not show the amount for completions. It looks very busy.

```python
g = sns.lmplot(y='cmp', # Variable 1.
               x='att', # Variable 2.
               data=df, # Data
               fit_reg=False, # If set to true, plots a regression line.
                ) 
g.set_ylabels("Completions")
g.set_xlabels("Attemps")
plt.title('Scatterplot: Completions by Attemps')
plt.show()
```

Pro: Easy to read.  Con: Does not have a Regression line

```python
g = sns.lmplot(y='cmp',
               x='att',
               data=df, # Data.
               #col='time',
               fit_reg=True,
               ci=False,
               size=5,
               scatter_kws={'alpha':0.4})
g.set_ylabels("Completions")
g.set_xlabels("Attemps")
plt.title('Scatterplot Regression: Completions by Attemps')

plt.show()
```

Pro: Has a regression line.  Con: none

```python
grid = sns.JointGrid(y='cmp', x='att', data=df, space=0, size=5, ratio=25)
g=grid.plot_joint(plt.scatter, color="g")
grid.plot_marginals(sns.rugplot, height=1, color="g")
plt.show()
```

Pro: Has marginal ticks.  Con: Does not have a Regression line

```python
y=df['cmp']
x=df['home_away']
g=sns.barplot(x, y)
g.set_ylabel("Completions")
plt.title('Completions: Home VS. Away')
plt.show()
```

Pros: Easy to read. Contains an error bar. Con: A lot of wasted space.

```python
# Set up the matplotlib figure

home=df[df['home_away'] == 'home']
away=df[df['home_away'] == 'away']

plt.hist(away['cmp'], alpha=.5, label='Away') 
plt.xlabel('Completions')
plt.title('Away Completions')

plt.hist(home['cmp'], alpha=.5, label='Home') 
plt.xlabel('Completions')
plt.title('Completions: Home VS. Away')

plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

Pro: There is a lot detail is in this plot. The amount of completions and the distributions are shown. Con: It is hard to read.

```python
g=sns.boxplot(x="home_away", y="cmp", data=df ,palette="PRGn")
sns.despine(offset=10, trim=True)
g.set_ylabel("Completions")
g.set_xlabel("")
plt.title('Completions: Home VS. Away')
plt.show()
```

Pro: The means for completions are shown. Cons : It does not show the amount for completions. The nuance of distributions is missing. 

```python
sns.set(style="darkgrid")

g = sns.FacetGrid(df, col="home_away", margin_titles=True)
bins = np.linspace(0, 60, 13)
g.map(plt.hist, "cmp", color="steelblue", bins=bins, lw=0)
plt.show()
```

Pro: There is a lot of detail in this plot. The amount of completions and the distributions are shown. Con: It's hard to read.

```python
sns.set(style="whitegrid")

g = sns.factorplot(x="home_away", y="cmp", data=df,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Completions")
g.set_xlabels("")
plt.title('Completions: Home VS. Away')
plt.show()
```

Pro: Clean and easy to read. Con: None
