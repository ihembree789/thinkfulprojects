```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
geometric_var = np.random.geometric(0.45, 100)
gumbel_var = np.random.gumbel(1, 1.1 , 100)
wald_var = np.random.wald(1, 2, 100)
poisson_var = np.random.poisson(3, 100)
weibull_var = np.random.weibull(1, 100)
exponential_var = np.random.exponential(2, 100)
```

```python
plt.hist(geometric_var)
plt.title('Geometric Distribution')
plt.axvline(geometric_var.mean(), color='black', linestyle='solid', linewidth=2)
plt.axvline(geometric_var.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
plt.hist(gumbel_var)
plt.title('Gumbel Distribution')
plt.axvline(gumbel_var.mean(), color='black', linestyle='solid', linewidth=2)
plt.axvline(gumbel_var.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
plt.hist(exponential_var)
plt.title('Exponential Distribution')
plt.axvline(exponential_var.mean(), color='black', linestyle='solid', linewidth=2)
plt.axvline(exponential_var.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
plt.hist(wald_var)
plt.title('Wald Distribution')
plt.axvline(wald_var.mean(), color='black', linestyle='solid', linewidth=2)
plt.axvline(wald_var.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
plt.hist(weibull_var)
plt.title('Weibull Distribution')
plt.axvline(weibull_var.mean(), color='black', linestyle='solid', linewidth=2)
plt.axvline(weibull_var.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
plt.hist(poisson_var)
plt.title('Poisson Distribution')
plt.axvline(poisson_var.mean(), color='black', linestyle='solid', linewidth=2)
plt.axvline(poisson_var.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
normal1 = np.random.normal(5,.5, 100)
normal2 = np.random.normal(10,1, 100)
sum_normal = normal1+normal2
mean = np.mean(sum_normal)
sd = np.std(sum_normal)
```

```python
plt.hist(sum_normal)
plt.axvline(x=mean,color='b')
plt.axvline(x=mean+sd,linestyle='dashed', color='r')
plt.axvline(x=mean-sd, linestyle='dashed', color='r')
plt.show()
```
