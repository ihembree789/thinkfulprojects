```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('Leading_Causes_of_Death.csv')
```

```python
df_us=df[df['State'] == "United States"]
df_us_can=df_us[df_us['Cause Name']=="Cancer"]
##print(df_us_can)
```

```python
plt.scatter(x=df_us_can["Year"],y=df_us_can["Deaths"])
plt.title("Cancer Deaths in U.S. by Year")
plt.ylim([548000, 600000])
plt.ylabel('Cancer Deaths')
plt.show()
```

__The scatter plot above shows the number of cancer deaths in the U.S. from 1999 to 2015. The total number of deaths appears to increase over time.__

```python
plt.figure(figsize=(10, 5))
#df_us=df[df['State'] == "United States"]
#df_us_can=df_us[df_us['Cause Name']=="Cancer"]



plt.subplot(1, 2, 1)
plt.plot(df_us_can['Age-adjusted Death Rate'], color='purple')
plt.ylabel('Age-adjusted Death Rate per 100,000')
plt.title('Death Rate from Caner in the U.S.')

plt.subplot(1, 2, 2)
plt.scatter(x = df_us_can['Year'], y = df_us_can['Age-adjusted Death Rate'], color='green')
plt.xlim([1998.5, 2015.5])
plt.ylabel('Age-adjusted Death Rate per 100,000')
plt.title('Death Rate from Caner in the U.S.')

plt.tight_layout()
plt.show()
```

__The subplot above shows two different plots with the same data. The first subplot is a line plot, and the second plot is a scatter plot. They both show age-adjusted death rate for cancer decreased in the U.S. from 1999 to 2015.__

```python
x=df_us_can['Age-adjusted Death Rate']

plt.hist(x, bins=30, color="green")
plt.title('Cancer Death Rates in the U.S. from 1999 to 2015')
plt.xlabel('Age-adjusted Death Rate per 100,000')
#plt.ylim([155, 202])
plt.show()
```

__This histogram plot shows the age-adjusted cancer death rate in the U.S. from 1999 to 2015. The range for the age-adjusted cancer death rate is roughly between 160 and 200 per 100,000. It also shows that, twice, the age-adjusted death rates are similar for different years.__

```python
plt.boxplot(x)
plt.title('Cancer Death Rates in the U.S. from 1999 to 2015')
plt.ylabel('Age-adjusted Death Rate per 100,000')
plt.show()
```

__This boxplot shows that age-adjusted death rate is divided evenly among the four different quartiles. It also shows the median is around 180.__
